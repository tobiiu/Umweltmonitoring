```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# CSV-Datei laden (Datum in Index setzen)
df = pd.read_csv("sensordaten.csv", index_col=1, parse_dates=True,)
```

```python
def plot_time_series(df, columns):
    """
    Zeichnet ein Liniendiagramm für eine oder mehrere Spalten eines DataFrames mit Zeitstempel im Index.

    Parameters:
    - df (pd.DataFrame): DataFrame mit datetime-Index
    - columns (list or str): Name(n) der Spalten, die geplottet werden sollen
    """
    if isinstance(columns, str):
        columns = [columns]

    # Nur vorhandene Spalten verwenden
    available_columns = [col for col in columns if col in df.columns]
    if not available_columns:
        print("⚠️ Keine der angegebenen Spalten im DataFrame vorhanden.")
        return

    plt.figure(figsize=(14, 6))
    for col in available_columns:
        plt.plot(df.index, df[col], label=col)

    plt.xlabel("Zeit")
    plt.ylabel("Messwert")
    plt.title("Zeitreihen-Diagramm")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Plot aufrufen
plot_time_series(df, df.columns[3:7])
```
